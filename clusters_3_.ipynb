# Importar las librerias y los algoritmos necesarios para la tarea

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, DBSCAN
from sklearn.datasets import make_blobs
```

# Generar datos aleatorios usando make_blobs

```python
np.random.seed(42)
n_puntos = 1000
centros = np.array([[0, 0], [1, 1], [2, 2]])
datos, _ = make_blobs(n_samples=n_puntos, centers=centros, cluster_std=0.2,random_state=42)
```

# Aplicar los modelos

## K-Means

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(datos)
etiquetas_kmeans = kmeans.labels_
```

## DBSCAN

```python
dbscan = DBSCAN(eps=0.34)
dbscan.fit(datos)
etiquetas_dbscan = dbscan.labels_
```

# Evaluaci√≥n de los modelos

```python
from sklearn import metrics
kmeans_metric = metrics.silhouette_score(datos,etiquetas_kmeans)
dbscan_metric = metrics.silhouette_score(datos,etiquetas_dbscan)
print("KMeans silhouette score = ", kmeans_metric)
print("DBSCAN silhouette score = ", dbscan_metric)
```

# Visualizar los clusters

```python
plt.figure(figsize=(5, 5))

# KMeans
plt.scatter(datos[:, 0], datos[:, 1], c=etiquetas_kmeans, cmap='viridis')
plt.title("KMeans")
plt.xlabel("Coordenada X")
plt.ylabel("Coordenada Y")

plt.show()
```

```python
# DBSCAN

plt.scatter(datos[:, 0], datos[:, 1], c=etiquetas_dbscan, cmap='viridis')
plt.title("DBSCAN")
plt.xlabel("Coordenada X")
plt.ylabel("Coordenada Y")

#plt.tight_layout()
plt.show()
```

   - Se especifica el n√∫mero de puntos (n_puntos), los centros de los clusters (centros), la desviaci√≥n est√°ndar de los clusters (cluster_std) y una semilla aleatoria (random_state) para la reproducibilidad.
   - La funci√≥n make_blobs devuelve un array con los datos y un array con las etiquetas de los clusters (que no se utilizan en este caso).
